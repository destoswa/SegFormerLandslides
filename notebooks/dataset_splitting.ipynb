# Dataset splitting

```python
import os
import shutil
import random
from tqdm import tqdm
```

```python
src_dataset= r"d:\Terranum_SD\99_Data\Landslide\data\Bern_glissements_spontane_shpfiles\split_images"
src_trainset = os.path.join(src_dataset, 'trainset')
src_valset = os.path.join(src_dataset, 'valset')
val_frac = 0.1
```

```python
os.makedirs(src_trainset, exist_ok=True)
os.makedirs(src_valset, exist_ok=True)

full_dataset = [x for x in os.listdir(src_dataset) if not 'mask' in x and os.path.splitext(x)[1].lower() in ['.jpg', '.png', '.tif']]
random.seed(42)
val_dataset = random.sample(full_dataset, int(val_frac * len(full_dataset)))
train_dataset = [x for x in full_dataset if x not in val_dataset]

# create training set
for _, img in tqdm(enumerate(train_dataset), total=len(train_dataset)):
    mask = img.replace('image', 'mask')
    shutil.copyfile(
        os.path.join(src_dataset, img),
        os.path.join(src_trainset, img)
    )
    shutil.copyfile(
        os.path.join(src_dataset, mask),
        os.path.join(src_trainset, mask)
    )
    
# create training set
for _, img in tqdm(enumerate(val_dataset), total=len(val_dataset)):
    mask = img.replace('image', 'mask')
    shutil.copyfile(
        os.path.join(src_dataset, img),
        os.path.join(src_valset, img)
    )
    shutil.copyfile(
        os.path.join(src_dataset, mask),
        os.path.join(src_valset, mask)
    )
```
