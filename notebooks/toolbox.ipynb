```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
from PIL import Image
```

### Convert RGB and bin

```python
def rgb_to_bin(src_in, src_out):
    img_rgb  = np.asarray(Image.open(src_in))[..., 0:3]
    img_bin = np.sum(img_rgb, axis=2)
    img_bin[img_bin != 0] = 1
    result = Image.fromarray(img_bin.astype(np.uint8))
    result.save(src_out)

def bin_to_rgb(src_in, src_out):
    img_bin  = np.asarray(Image.open(src_in))
    img_rgb = np.zeros((img_bin.shape[0], img_bin.shape[1], 3), dtype=np.uint8)
    img_rgb[img_bin == 1] = 255
    result = Image.fromarray(img_rgb, mode="RGB")
    result.save(src_out)
```

```python
#test
src_in = r"D:\GitHubProjects\Terranum_repo\LandSlides\segformerlandslides\data\dataset_Bern\test\1998_0_7.tif"
src_out = r"D:\GitHubProjects\Terranum_repo\LandSlides\segformerlandslides\data\dataset_Bern\test\1998_0_7_rgb.tif"
bin_to_rgb(src_in, src_out)
```

```python
# Convert everything in folder from bin to rgb
src_bin_folder = r"D:\GitHubProjects\Terranum_repo\LandSlides\segformerlandslides\results\training\20251217_155250_50_epochs_Bern_v2_da_splitsets\logs\best_preds\reconstructed"
src_rgb_folder = os.path.join(os.path.dirname(src_bin_folder), os.path.basename(src_bin_folder) + "_rgb")
os.makedirs(src_rgb_folder, exist_ok=True)
for _, filename in tqdm(enumerate(os.listdir(src_bin_folder)), total=len(os.listdir(src_bin_folder))):
    bin_to_rgb(
        os.path.join(src_bin_folder, filename),
        os.path.join(src_rgb_folder, filename),
        )
```

```python
# Convert everything in folder from rgb to bin
src_rgb_folder = r"D:\GitHubProjects\Terranum_repo\LandSlides\segformerlandslides\data\dataset_Longxi\labels"
src_bin_folder = r"D:\GitHubProjects\Terranum_repo\LandSlides\segformerlandslides\data\dataset_Longxi\masks"
os.makedirs(src_bin_folder)
for _, filename in tqdm(enumerate(os.listdir(src_bin_folder)), total=len(os.listdir(src_bin_folder))):
    rgb_to_bin(
        os.path.join(src_rgb_folder, filename),
        os.path.join(src_bin_folder, filename),
        )
```
